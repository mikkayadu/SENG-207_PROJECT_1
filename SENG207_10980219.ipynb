## Heatmap to show correlation between attributes

```python

plt.figure(figsize=(15, 15))
sns.heatmap( df.drop('ID', axis=1).corr(), annot=True)
```

## Inference from heatmap
1. The higher the fixed acidity of the beverage , the lower the pH. This means fixed acidity and pH has a **strong negative** correlation
2. Also, it can be noticed that fixed acidity has a **strong positive** correlation with the citric acid content.
3. The ` content of a beverage` has **strong correlation** with the `free sulfur dioxide` .
4. `fixed acidity` has a **strong** ** positive correlation** with `citric acid`.
5. `alcohol` has a **medium negative correlation** with `density`. This means that averagely, drinks with high alcohol content have low density.
6. `pH` has a **medium negative correlation** with `citric acid`. This means that averagely, drinks with high Ph have low citric acid content. And this is because the higher the pH , the more basic the drink is.
7. `density` has a **strong postive correlation** with `fixed acidity`. This means drinks with high acidity are likely to have high density.

```python
fig, axs = plt.subplots(ncols=5,nrows=3, figsize=(15, 10))

sns.kdeplot(data=df, x='fixed acidity', ax=axs[0,0])
sns.kdeplot(data=df, x='volatile acidity', ax=axs[0, 1])
sns.kdeplot(data=df, x='citric acid', ax=axs[0,2])
sns.kdeplot(data=df, x='residual sugar', ax=axs[0,3])
sns.kdeplot(data=df, x='chlorides', ax=axs[0,4])
sns.kdeplot(data=df, x='free sulfur dioxide', ax=axs[1,0])
sns.kdeplot(data=df, x='total sulfur dioxide', ax=axs[1, 1])
sns.kdeplot(data=df, x='density', ax=axs[1,2])
sns.kdeplot(data=df, x='pH', ax=axs[1,3])
sns.kdeplot(data=df, x='sulphates', ax=axs[1,4])
sns.kdeplot(data=df, x='alcohol', ax=axs[2, 0])
sns.kdeplot(data=df, x='quality', ax=axs[2, 1])
```
